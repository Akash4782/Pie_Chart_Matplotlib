Pie Chart/Plot

```python
import matplotlib.pyplot as plt 

x = [10,20,30,40]
y = ["C" , "C++" , "Java" , "Python"]
plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C" , "C++" , "Java" , "Python"]
plt.pie(x , labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C" , "C++" , "Java" , "Python"]
ex = [0.2 , 0.0 , 0.0 , 0.0]
plt.pie(x , labels = y , explode = ex)
plt.show()
```

```python
x = [10,20,30,40]
y = ["C" , "C++" , "Java" , "Python"]
ex = [0.2 , 0.0 , 0.0 , 0.0]
c = ["r","g","b","y"]
plt.pie(x , labels = y , explode = ex , colors = c)
plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["C" , "C++" , "Java" , "Python"]
c = ["r","b","g","y"]
plt.pie(x , labels = y , radius = 1)
cr = plt.Circle(xy =(0,0), radius = 0.7 , facecolor = "w")
plt.gca().add_artist(cr)
plt.show()
```
